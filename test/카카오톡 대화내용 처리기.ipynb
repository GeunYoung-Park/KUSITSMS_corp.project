```python
# try, exceptë¡œ ì˜ˆì™¸ì²˜ë¦¬í•˜ê³ , forë¬¸ì„ ì´ìš©í•´ì„œ ëŒ€í™”ë‚´ìš© listì— ë‹´ê¸°
content_all = []
for i in data['Talk_2021.3.25 21:18-1.txt']:
    content = i.split(',',2)
    try:
        content2 = content[1].split(':',2)
        content_all.append(content[0] + ',' + content2[0] + ',' + content2[1])
    except:
        continue
```

```python
content_all
```

```python
p = []
pp = []
ppp = []
for ii in content_all:
    iii = ii.split(',',3)
    p.append(iii[0])
    pp.append(iii[1])
    ppp.append(iii[2])
```

```python
content_df = pd.DataFrame(data = p, columns = ['date'])
```

```python
content_df['name'] = pp
```

```python
content_df['content'] = ppp
```

```python
content_df
```
